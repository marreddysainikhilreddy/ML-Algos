```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
digit = load_digits()
```

```
len(digit.data)
```

```
len(digit.target)
```

```
data = pd.DataFrame(digit.data, digit.target)
data
```

```
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(data, digit.target, train_size=0.7, random_state=42)
```

```
X_train
```

```
X_test
```

```
from sklearn.svm import SVC
model = SVC()
```

```
model.fit(X_train, y_train)
```

```
model.score(X_test, y_test)
```

```
model1 = SVC(kernel="linear")
```

```
model1.fit(X_train, y_train)
```
